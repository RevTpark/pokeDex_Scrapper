```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://pokemondb.net/type"
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.109 Safari/537.36"}
r = requests.get(url, headers=headers)
r
```

```python
soup = BeautifulSoup(r.content, "html.parser")
type_list = soup.find("table", {"class": "type-table"})
type_list
```

```python
types_order = []
for types in type_list.find_all("th")[1:]:
    try:
        types_order += [types.find("a")['title']]
    except:
        pass
types_order
```

```python
type_list.find_all('tr')[1].find_all('td')[0]
```

```python
attack = { ty: {} for ty in types_order }
defense = {ty: {} for ty in types_order }
```

```python
for ty in type_list.find_all('tr'):
    for each in ty.find_all('td'):
        li = each['title'].split('→')
        before = li[0].strip()
        after = li[1].split("=")[0].strip()
        if each.text == "":
            effect = "1"
        else:
            effect = each.text
        attack[before][after] = effect
        defense[after][before] = effect
```

```python
attack, defense
```

```python
attack_df = pd.DataFrame(attack)
defense_df = pd.DataFrame(defense)
```

```python
attack_df.to_csv("attack.csv")
defense_df.to_csv("defense.csv")
```
