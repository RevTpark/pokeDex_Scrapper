```python
data_dir1 = "/content/drive/MyDrive/pokemon_ml/mixed_dataset"
data_dir2 = "/content/drive/MyDrive/pokemon_ml/pokemon"
```

```python
import os
dir1 = set(os.listdir(data_dir1))
dir2 = set(os.listdir(data_dir2))
# dir1.difference(dir2)
```

```python
import shutil
# shutil.copy("/content/drive/MyDrive/pokemon_ml/pokemon/abra/abra_0.png", "/content/drive/MyDrive/pokemon_ml/mixed_dataset/Abra")
```

```python
for dir in dir2:
  if dir in ['nidoran_male', 'nidoran_female']:
      continue
  if dir == 'mr.mime':
      copy_to = data_dir1+"/MrMime"
  else:
      copy_to = data_dir1+"/"+dir.capitalize()
  path1 = data_dir2+"/"+dir
  for filename in os.listdir(path1):
      temp_path = path1+"/"+filename
      shutil.copy(temp_path, copy_to)
```

```python
for dir in dir1:
  path2 = data_dir1+'/'+dir
  cnt = 0
  for filename in os.listdir(path2):
    cnt += 1
  print(cnt)
```
