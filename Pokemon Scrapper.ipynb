```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://pokemondb.net/pokedex/all"
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.109 Safari/537.36"}
r = requests.get(url, headers=headers)
r
```

```python
soup = BeautifulSoup(r.content, "html.parser")
pokemon_list = soup.find("table", {"id": "pokedex"}).find_all('tr')
pokemon_list
```

```python
pokedexbase = {}

for pokemon in pokemon_list[1:]:
    poke_id = int(pokemon.find('span', {"class": "infocard-cell-data"}).text)
    name = pokemon.find('a', {"class": "ent-name"}).text
    poke_types = pokemon.find_all('a', {"class": "type-icon"})
    primary_type = poke_types[0].text
    secondary_type = None
    if len(poke_types) != 1:
        secondary_type = poke_types[1].text
    
    is_mega = False
    if pokemon.find('small', {"class": "text-muted"}):
        name = pokemon.find('small', {"class": "text-muted"}).text
        is_mega = True
    
    if pokedexbase.get(poke_id, None) == None:
        pokedexbase[poke_id] = name
#     print(poke_id, name, primary_type, secondary_type, is_mega)
print(pokedexbase)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://projectpokemon.org/home/docs/spriteindex_148/3d-models-generation-1-pok%C3%A9mon-r90/"
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.109 Safari/537.36"}
r = requests.get(url, headers=headers)
r
```

```python
soup = BeautifulSoup(r.content, "html.parser")
pokemon_list = soup.find("table")
pokemon_list
```

```python
import re
poke_data =  {}
for row in pokemon_list.find_all('td'):
    link = row.find('img')
    if re.search('normal-sprite', link['src']) and not re.search('mega', link['src']) and not re.search('alola', link['src']) and not re.search('pikachu-', link['src']):
        
        poke_data[link['alt'].rstrip('.gif')] = link['src']
poke_data
```

```python
from PIL import Image
from PIL import GifImagePlugin
import urllib.request
import os

parent_dir = 'C:/Users/tanis/Jupyter Projects/Pokemon Scrapper/pokemon'


for mon, url in poke_data.items():
    urllib.request.urlretrieve(url,"temp1.gif")
    imageObject = Image.open('temp1.gif')
    
    directory = mon
    path = os.path.join(parent_dir, directory)
    if not os.path.exists(path):
        os.mkdir(path)
    
    for frame in range(0,imageObject.n_frames):
        imageObject.seek(frame)
        imageObject.save(f"pokemon/{mon}/{mon}_{frame}.png", "PNG")
    
```
