```python
def get_desc_for(pokemon):
    if pokemon == 'Nidoran-f': pokemon = 'nidoran-female'
    elif pokemon == 'Nidoran-m': pokemon = 'nidoran-male'
    
    url = f"https://www.pokemon.com/us/pokedex/{pokemon}"
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, "html.parser")
    desc = soup.find('p', {"class": "version-x active"}) or  soup.find('p', {"class": "version-y active"})
    desc = desc.text.lstrip("\n ").rstrip("\n ")
    return desc
```

```python
test_dict = {}
for pokemon in class_names:
    desc = get_desc_for(pokemon)
    test_dict[pokemon] = desc
    time.sleep(0.5)
test_dict
```

```python
dict_list = [{ "Pokemon": k, "Description": v } for k, v in test_dict.items() ]
dict_list
```
