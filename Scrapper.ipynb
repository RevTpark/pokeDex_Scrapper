```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://pokemon.fandom.com/wiki/Category:Generation_I_Pok%C3%A9mon"
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.109 Safari/537.36"}
r = requests.get(url, headers=headers)
r
```

```python
soup = BeautifulSoup(r.content, "html.parser")
pokemon_list = soup.find_all("img", {'class': "thumbimage"})
pokemon_list
```

```python
new_data = {}
for poke in pokemon_list:
  if not new_data.get(poke['alt'].lower()):
    new_data[poke['alt'].lower()] = poke['src']
new_data
```

```python
new_data['venonat'] = new_data['048venonat']
new_data['venomoth'] = new_data['049venomoth']
new_data.pop('048venonat')
new_data.pop('049venomoth')
```

```python
url = "https://pokemondb.net/pokedex/national"
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.109 Safari/537.36"}
r = requests.get(url, headers=headers)
r
```

```python
soup = BeautifulSoup(r.content, "html.parser")
pokemon_list = soup.find("div", {"class": "infocard-list"}).find_all("span", {"class": "img-fixed"})
pokemon_list
```

```python
new_data3 = {}
for poke in pokemon_list:
  new_data3[poke['data-alt'].split(" ")[0].lower()] = poke['data-src']
new_data3
```

```python
new_data3["mr.mime"] = new_data3['mr.']
new_data3["farfetchd"] = new_data3["farfetch'd"]
new_data3['nidoran_female'] = new_data3["nidoran♀"]
new_data3['nidoran_male'] = new_data3['nidoran♂']
new_data3.pop('mr.')
new_data3.pop("farfetch'd")
new_data3.pop("nidoran♀")
new_data3.pop("nidoran♂")
```

```python
for item in constant:
  if not new_data3.get(item):
    print(item)
```

```python
for poke in constant:
  if new_data3.get(poke):
    opener = urllib.request.build_opener()
    opener.addheaders = [(k, v) for k, v in headers.items()]
    urllib.request.install_opener(opener)
    urllib.request.urlretrieve(new_data3[poke], "temp.png")
    imageObject = Image.open('temp.png')
    imageObject.save(f"/content/drive/MyDrive/pokemon_ml/pokemon/{poke}/{poke}_new_im3.png")
```
